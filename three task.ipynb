```python
import folium
from folium import Circle

# Create a base map centered on Bengaluru
map_center = [12.9716, 77.5946]
density_map = folium.Map(location=map_center, zoom_start=12)

# Sample data: [Latitude, Longitude, Density Value]
# Density here can represent number of restaurants in the area or people etc.
location_density = [
    [12.9750, 77.6050, 50],
    [12.9710, 77.5950, 80],
    [12.9735, 77.6110, 30],
    [12.9690, 77.5940, 100],
    [12.9765, 77.6000, 60],
    [12.9800, 77.6200, 90],
    [12.9650, 77.5850, 40],
    [12.9600, 77.5900, 70],
    [12.9700, 77.6100, 20],
    [12.9720, 77.5980, 85]
]

# Add circle markers based on density
for lat, lon, density in location_density:
    folium.Circle(
        location=[lat, lon],
        radius=density * 5,  # Increase size based on density
        color='crimson',
        fill=True,
        fill_opacity=0.5,
        popup=f"Density: {density}"
    ).add_to(density_map)

# Display the map
density_map
```
